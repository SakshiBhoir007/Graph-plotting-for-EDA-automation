##### Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

##### Importing the datasets


```python
# Importing cars dataset
cars = pd.read_csv("cars.csv")

# Importing Iris dataset
iris = pd.read_csv("Iris.csv")
```

**About the function** 

Here we have created a function called "graphs". This function will find out the datatype of each column (which are passed by users or default: all)  will categorise them on the based of numerical or categorical column and will return the graph according to the their datatypes. Histogram and boxplot for numerical datatype column and Barplot for categorical columns.

```python
def graphs(data,cols = [],directory= os.getcwd()):
    
    """
    data:-
    The data variable is the compulsory input for the function. The user must pass
    a dataframe in this function to be able to execute this function.
    """

    """ 
    Cols:- 
    So if the user has not passed any input for the column names then,
    the length of column input will be zero. therefore, we are passing
    the "if" statement stating that if the  the length of column of columns is zero
    then the column variable will be the all the columns of the data which we can get
    data.columns.
    """
    
    """
    Directory:-
    By default the current path of the directory will be the present working of the file.
    which can be fetched by "os.getcwd()". Otherwise the user can specify the directory by
    their choice and the image file will be saved in that specified directory
    """
    
    # checking if the user has passed any arguments cols or not. 
    # if no then the default values would be the all the columns of the dataset
    if len(cols)==0: 
        cols = data.columns

        
    # checking columns datatype: if it is numerical or categorical
    for column in cols:       
        # if column datatype is numeric
        if data[column].dtype != "object":       
            plt.subplot(2,1,1) # Plot subplot(Row, columns, chart no.)
            plt.subplots_adjust(hspace=1)  # height of padding between subplots
            plt.hist(data[column],color="aqua",edgecolor="black") # Plot histogram for that column 
            plt.title("Histogram of {}".format(column), fontweight="bold") # Plot title
            plt.xlabel(column)  # here X axis label is the name of the plotted column
            plt.ylabel("frequency") # y label

            plt.subplot(2,1,2)  # Plot subplot(Row, columns, chart no)
            plt.subplots_adjust(hspace=1) #  height of padding between subplots
            plt.boxplot(data[column],vert=False,  patch_artist=True)
            plt.title("Box Plot of {}".format(column),fontweight="bold") # Plot title
            plt.ylabel(column) # here Y axis label is the name of the plotted column
            plt.savefig("{}/{}.png".format(directory,column)) # save the figure in directory with the column name which is plotted
            plt.show()
            
        # if columns datatype is categorical 
        else:       
            data[column].value_counts().plot(kind='barh',color="moccasin")  # plot Bar plot 
            plt.title("Bar Plot of {}".format(column),fontweight="bold") # plot title 
            plt.xlabel(column) # here X axis label is the name of the plotted column
            plt.savefig("{}/{}.png".format(directory,column)) # save the figure in directory with the column name which is plotted
            plt.show()


```

###### Executing the function with 1 arguments i.e. cars dataset

```python
graphs(cars)
```

###### Executing the function with two arguements i.e. cars dataset and limited columns ["MPG","Origin"]

```python
graphs(cars,cols = ["MPG","Origin"])
```
